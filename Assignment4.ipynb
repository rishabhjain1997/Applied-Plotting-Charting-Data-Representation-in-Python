```python
import numpy as np
import pandas as pd
state_data = pd.read_excel('India_state_economic_data.XLSX', header = 3, index_col = 1, skipfooter = 2)
state_data.drop(state_data.loc[state_data.index.str.contains(r'Base|^1$')].index,axis = 0, inplace = True)
#state_data.loc[state_data.index.str.contains(r'Base')].index
state_data.columns
```

```python
state_data = state_data[['Bihar','Madhya Pradesh', 'Rajasthan','Uttar Pradesh']]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
state_data.index
```

```python
plt.figure(figsize = (18,10));
plt.plot(state_data.index, state_data['Bihar'], state_data.index, state_data['Madhya Pradesh'], state_data.index, state_data['Rajasthan'], 
         state_data.index, state_data['Uttar Pradesh']);
```

```python
#state_data.index.replace(to_replace=r'(\d{4})-\d\d',value=r"\1", regex=True, inplace=True)
state_data.reset_index(inplace = True)
state_data['index'].replace(to_replace=r'(\d{4})-\d\d',value=r"\1", regex=True, inplace=True)
state_data.set_index('index', inplace = True)
state_data
```

```python
import seaborn as sns
plt.style.use('seaborn')
plt.figure(figsize = (18,10));
plt.plot(state_data.index, state_data['Bihar'], label = "Bihar")
plt.plot(state_data.index, state_data['Madhya Pradesh'], label = "Madhya Pradesh")
plt.plot(state_data.index, state_data['Rajasthan'], label = "Rajasthan")
plt.plot(state_data.index, state_data['Uttar Pradesh'], label = "Uttar Pradesh")
plt.legend()
left_side = plt.gca().spines["bottom"]
left_side.set_visible(True) 
plt.gca().tick_params(axis="x", direction="out", length=6, width=2, color="#d3d3d3")
plt.gca().tick_params(axis="y", direction="out", length=6, width=2, color="#d3d3d3")
plt.gca().xaxis.label.set_alpha(0.1)
plt.gca().yaxis.label.set_color("#d3d3d3")
plt.title('PER CAPITA NET STATE DOMESTIC PRODUCT- STATE-WISE ')
plt.ylabel('Per-capita NDSP at current prices in INR', color = 'black')
plt.grid(False)
plt.show()
```
